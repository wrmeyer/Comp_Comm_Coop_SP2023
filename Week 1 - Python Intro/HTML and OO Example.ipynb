```python
import requests
from bs4 import BeautifulSoup
```

```python
# Make a request to the Wikipedia page
url = "https://en.wikipedia.org/wiki/Main_Page"
response = requests.get(url)
```

```python
#examine 
dir(response)
```

```python
response.text
```

```python
# Parse the HTML of the page
print("begin")
soup = BeautifulSoup(response.text, 'html.parser')
print("end")
```

```python
# Find all 'a' tags
a_tags = soup.find_all('a')[0:50]
```

```python
# Print out the 'a' tags
for thing in a_tags:
    print(thing)
```

Now we have a list of dictionaries

```python
my_list = []
x = 1
while x < 10:
    my_list.append(x)
    x += 1
    
print(my_list)
```

```python
simple_urls = []

for a_tag in a_tags:
    simple_urls.append(a_tag['href'])

for url in simple_urls:
    print(url)
```

```python
for url in simple_urls:
    if "https:" not in url:
        print("https://en.wikipedia.org" + url)
        
hyper_links = []
for url in simple_urls:
    if "https:" not in url:
        hyper_links.append("https://en.wikipedia.org" + url)
    else:
        hyper_links.append(url)
```

```python
with open('readme.txt', 'w') as file:
    for link in hyper_links:
        file.write(link)
        file.write('\n')
```

```python
with open("readme.txt", "r") as file:
    lines = file.readlines()
    
print(lines)
```
